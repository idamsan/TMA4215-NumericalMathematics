## Problem 1

We consider the generalized eigenvalue problem: Let $A, B \in \mathbb FR^{n\times n}$ be symmetric positive definite matrices. Then, a pair $(\lambda, \mathbf x) \in \mathbb R\times \mathbb R^n$ is a generalized  eigenvalue $\lambda$ with generalized eigenvector $\mathbf x$ if

$$
A\mathbf x = \lambda B\mathbf x
$$

holds.

Prove that the smallest and largest eigenvalues $\lambda_{\mathrm{min}},\lambda_{\mathrm{max}}$ of the generalized eigenvalues can be computed as

$$
\lambda_{\mathrm{max}}
  =
    \min_{\substack{\mathbf{x}\in\mathbb R^n \\ \mathbf{x}\neq\mathbf{0}}}
    \frac{\mathbf{x}^\mathrm{T} A\mathbf{x}}
    {\mathbf{x}^\mathrm{T} B \mathbf{x}}
  \qquad\text{and}\qquad
    \lambda_{\mathrm{max}}
  =
    \max_{\substack{\mathbf{x}\in\mathbb R^n \\ \mathbf{x}\neq\mathbf{0}}}
    \frac{\mathbf{x}^\mathrm{T} A\mathbf{x}}
    {\mathbf{x}^\mathrm{T} B \mathbf{x}}.
$$

_Hint_: Use the Rayleigh quotient (from Project 1) and the Cholesky decomposition.

Looking at the hint it seems reasonable to begin with $\textbf{F}(\textbf{x}) = \frac{\textbf{x}^TA\textbf{x}}{\textbf{x}^TB\textbf{x}}$ and compute the gradient of this, to hopefully have eigenvalues show up. We know both from project 1 and assignment 2 that 

$\nabla (\textbf{x}^TA\textbf{x}) = (A^T+A)\textbf{x} = 2A\textbf{x}$, for symmetric matrix A. 

Using the quotient rule we then get that 

$\nabla (\frac{\textbf{x}^TA\textbf{x}}{\textbf{x}^TB\textbf{x}}) = \frac{2A\textbf{x}(\textbf{x}^TB\textbf{x})-\textbf{x}^TA\textbf{x}(2B\textbf{x})}{(\textbf{x}^TB\textbf{x})^2}$.

Rearranging gives 

$\frac{2}{\textbf{x}^TB\textbf{x}}\left(A-\textbf{F}(\textbf{x})B\right)\textbf{x}$. 

Stationary points when $\nabla\textbf{F}(\textbf{x}) = 0$:

$\implies$

$(A-\textbf{F}(\textbf{x})B)\textbf{x}=0$

$A\textbf{x}=\textbf{F}(\textbf{x})B\textbf{x}$.

Since we have from the problem text that $A\textbf{x}=\lambda B \textbf{x}$, we see that $\textbf{F}(\textbf{x})=\frac{\textbf{x}^TA\textbf{x}}{\textbf{x}^TB\textbf{x}}=\lambda.$

From this it is easy to see that $\lambda_{min}$ and $\lambda_{max}$ follows from the expressions in the problem text.

```python
def F(x):
    return (3*x**3-3*x+3)**3

x_vals=np.array((0,1/12,2/12,3/12,4/12,5/12,6/12,7/12,8/12,9/12,10/12,11/12,1))
```

```python
def dividedDifferences(x, y): #x, y arrays
    
    m = len(x)
    x = np.copy(x) #copy to not change the initial x
    f = np.copy(y) #copy to not change the initial y, and get the right length, and get f[x_i]
    
    #filling f with all the divided differences
    for k in range(1, m):
        f[k:m] = (f[k:m] - f[k - 1])/(x[k:m] - x[k - 1])

    return f

# x_data, y_data same as x and y in dividedDifferences(x,y)
# x is the point you want to evaluate
def newton(x_data, y_data, x):
    a = dividedDifferences(x_data, y_data)
    n = len(x_data) - 1  # Degree of polynomial
    p = a[n]
    for k in range(1, n + 1):
        p = a[n - k] + (x - x_data[n - k])*p

    return p

print("The value p_12(2) is",newton(x_vals,F(x_vals),2))
```
